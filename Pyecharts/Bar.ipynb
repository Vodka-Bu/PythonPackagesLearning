# 柱状图

柱状/条形图，通过柱形的高度/条形的宽度来表现数据的大小。

## 特殊配置项

相对比较简单，并且之前通用配置的时候用了很多了

**name -> str**  
图例名称，没什么问题   

**x_axis -> list**  
x 坐标轴数据， 一般只有一个  


**y_axis -> list**  
y 坐标轴数据，每一个 name 可以有一组y_axis  

**is_stack -> bool**  
数据堆叠，同个类目轴上系列配置相同的 stack 值可以堆叠放置  

**bar_category_gap -> int/str**  
类目轴的柱状距离，当设置为 0 时柱状是紧挨着（直方图类型），默认为 '20%'  

## 基本示例

```python
from pyecharts import Bar

x_axis = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
y_axis_1 = [5, 20, 36, 10, 75, 90]
y_axis_2 = [10, 25, 8, 60, 20, 80]
```

```python
bar = Bar('普通示例')
bar.add("商品A", x_axis, y_axis_1)
bar
```

## 堆叠图

```python
# 需要对所有要 stack 的数据，都加上 is_stack = True
bar = Bar('普通示例')
bar.add("商品A", x_axis, y_axis_1, is_stack = True)\
   .add("商品B", x_axis, y_axis_2, is_stack = True)
bar
```

## 加标注

```python
bar = Bar("标记线和标记点示例")
# 标注可以用 raw 格式输入自己想要的各种点
bar.add("商家A", x_axis, y_axis_1, mark_point=["average",{"coord": [4, 75], "name": "高跟鞋"}])
bar.add("商家B", x_axis, y_axis_2, mark_line=["min", "max"])
bar
```

## 坐标轴交换

```python
bar = Bar("标记线和标记点示例")
# 交换以后 x_axis 名字会消失，但是相同配置在线上没有该问题
bar.add("商家A", x_axis, y_axis_1)
bar.add("商家B", x_axis, y_axis_2, is_convert = True)
bar
```
